3- crie um algoritmo que receba o nome do aluno, e 3 notas. depois calcule a média das notas e imprima na tela:
o aluno (aluno) está com (media) de media!

```python
#entradas
nome_aluno = str(input("Digite o nome do aluno: "))
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

#processamento
media = (nota1 + nota2 + nota3) / 3

#saidas
print(f"O aluno {nome_aluno} está com {media:.2f} de media!")
```

Crie um algoritmo que receba um número do usuário e imprima a tabuada deste número de 0 até 10.


```python
numero = int(input("Digite um número para ver a tabuada 5: "))

for i in range(11):
  resultado = numero * i
  print(f"{numero} x {i} = {resultado}")
```

##While - enquanto

Repete uma estrutura até ela ser atingida

```python
numero = 0

while(numero < 10):
  print("menor que 10")
  numero = int(input("digete um numero:"))

print("Maior ou igual a 10")
```

JOGO DA ADIVINHAÇÃO


```python
aleatorio = rd.randint(0, 100)

print("===================================================")
print("             JOGO DE ADIVINHAÇÃO                   ")
print("===================================================")
print("  Tente adivinhar o número que estou pensando!     ")
print("===================================================")

tentativa = 10


while (tentativa > 0):
  numero = int (input("digite um numero:"))
  if (numero == aleatorio):
    print("parabéns ! Você acertou!")
    break
  elif (numero > aleatorio):
    print ("digite um número menor")
  else:
    print("Digite um número maior:")
    tentativa -= 1
    if (tentativa == 0):
      print(" game over")
```
